```python
class Stack:
    def __init__(self, max_num):
        self.max = max_num
        self.head = None
        self.tail = None
        self.number_of_nodes = 0
    
    def pop(self):
        if self.number_of_nodes == 1:
            node = self.tail
            return (False, node)
        else:
            self.number_of_nodes -= 1
            node = self.tail
            self.tail = self.tail.pre
            self.tail.next = None
            return (True, node)
    
    def remove_bottom(self):
        self.number_of_nodes -= 1
        head = self.head
        self.head = self.head.next
        
        
        if self.number_of_nodes == 0:
            return (False, head)
        else:
            self.head.pre = None
            return (True, head)
    
    def stack_print(self):  
        node = self.tail
        while node:
            print(node.data)
            node = node.pre
            
    def connect(self, pre, post):
        post.pre = pre
        pre.next = post
        
    def push(self, data):
        new_node = Node(data)
        if not self.tail:
            self.head = new_node
            self.tail = new_node
            self.number_of_nodes += 1
            return True
        else:
            if self.number_of_nodes == self.max:
                return False
            else:
                self.number_of_nodes += 1
                self.connect(self.tail, new_node)
                self.tail = new_node 
                return True
    
```

```python
class SetOfStack:
    def __init__(self, max_num):
        self.max = max_num
        self.list = []
    
    def push(self, data):
        if len(self.list) == 0:
            new_stack = Stack(self.max)
            new_stack.push(data)
            self.list.append(new_stack)
        else:
            last_stack = self.list[-1]
            pushed = last_stack.push(data)
            if not pushed:
                new_stack = Stack(self.max)
                new_stack.push(data)
                self.list.append(new_stack)

    def pop(self):
        if len(self.list) == 0:
            raise Exception("No stack in the set")
        else:
            last_stack = self.list[-1]
            result, node = last_stack.pop()
            if not result:
                self.list.pop()
            return node
        
    #solution one: no rolling over when pop_at(index) is called, leaving some stacks not being at 
    #full capacity is okie
    def popAt(self, index):
        if index >= len(self.list):
            raise Exception("index out of range")
        else:
            stack = self.list[index]
            result, node = stack.pop()
            if not result:
                self.list.pop(index)
            return node
```

```python
#solution two: rolling over when pop_at(index) is called, leaving all stacks at its full capacity except the last one

class SetOfStack:
    def __init__(self, max_num):
        self.max = max_num
        self.list = []
    
    def push(self, data):
        if len(self.list) == 0:
            new_stack = Stack(self.max)
            new_stack.push(data)
            self.list.append(new_stack)
        else:
            last_stack = self.list[-1]
            pushed = last_stack.push(data)
            if not pushed:
                new_stack = Stack(self.max)
                new_stack.push(data)
                self.list.append(new_stack)

    def pop(self):
        if len(self.list) == 0:
            raise Exception("No stack in the set")
        else:
            last_stack = self.list[-1]
            result, node = last_stack.pop()
            if not result:
                self.list.pop()
            return node
        
    #solution two: rolling over when pop_at(index) is called, leaving all stacks at its full 
    #capacity except the last one
    def popAt(self, index):
        if index >= len(self.list):
            raise Exception("index out of range")
        else:
            
            stack = self.list[index]
            result, node = stack.pop()
            if not result:
                self.list.pop(index)
                return node
            else:   
                return_value = node
                while index+1 < len(self.list):
                    print("len(self.list)", len(self.list))
                    stack = self.list[index]
                    result, node = self.list[index+1].remove_bottom()
                    
                    stack.push(node.data)
                    if not result:
                        self.list.pop(index+1)
                        break
                    index += 1
                return return_value
                
```

```python
class Tests(unittest.TestCase):
    def test_stacks(self):
        stacks = SetOfStack(5)
        for i in range(35):
            stacks.push(i)
        lst = []
        for _ in range(35):
            lst.append(stacks.pop())
        result = [i.data for i in lst]

        self.assertEqual(result, list(reversed(range(35))))

    def test_pop_at(self):
        stacks = SetOfStack(5)
        for i in range(35):
            stacks.push(i)
        lst = []
        for _ in range(31):
            lst.append(stacks.popAt(0).data)
        self.assertEqual(lst, list(range(4, 35)))

a = Tests()
a.test_stacks()
a.test_pop_at()
```
